```python
# match_id로 matchtimeline, match 데이터 읽어오는 함수
def open_json_file(match_timeline_dir_path,match_dir_path, match_id): 

    match_timeline_path = os.path.join(match_timeline_dir_path, f"{match_id}.json")
    match_path = os.path.join(match_dir_path,f"{match_id}.json")

    with open(match_timeline_path,'r') as f:
        timeline_data = json.load(f)
    with open(match_path,'r') as f:
        info_data = json.load(f)
    return timeline_data, info_data 

def get_match_id_list(match_id_list_path):
    df = pd.read_csv(match_id_list_path)
    match_id_list = df['match_id'].tolist()
    return match_id_list
```

```python
building_kill = []
champion_kill = []
champion_special_kill = []
elite_monster_kill = []
game_end = []
item_destroyed = []
item_purchased = []
item_sold = []
level_up = []
objective_bounty_prestart=[]
skill_level_up = []
turret_plate_destroyed = []
ward_kill = []
ward_placed = []
for i in range(len(event_data)):
    for j in range(len(event_data[i]['events'])):
        event = event_data[i]['events'][j]
        if event['type'] == 'BUILDING_KILL':
            building_kill.append({
                'match_id': match_id,
                'assistingParticipantIds': int("".join(map(str, event['assistingParticipantIds']))) if 'assistingParticipantIds' in event else 0,
                'bounty' : event['bounty'],
                'buildingType':event['buildingType'],
                'killerId':event['killerId'],
                'laneType':event['laneType'],
                'teamId':event['teamId'],
                'timestamp':event['timestamp'],
                'towerType':event['towerType'],
            })
        elif event['type'] == 'CHAMPION_KILL':
            champion_kill.append({
                'match_id': match_id,
                'assistingParticipantIds': int("".join(map(str, event['assistingParticipantIds']))) if 'assistingParticipantIds' in event else 0,
                'bounty': event['bounty'],
                'killStreakLength': event['killStreakLength'],
                'killerId': event['killerId'],
                'shutdownBounty': event['shutdownBounty'],
                'timestamp': event['timestamp'],
                'victimId': event['victimId']
            })
        elif event['type'] == 'CHAMPION_SPECIAL_KILL':
            champion_special_kill.append({
                'match_id':match_id,
                'killType': event['killType'],
                'killerId': event['killerId'],
                'multiKillLength': event['multiKillLength'] if 'multiKillLength' in event else 0,
                'timestamp': event['timestamp']
            })
        elif event['type'] == 'ELITE_MONSTER_KILL':
            elite_monster_kill.append({
                'match_id' : match_id,
                'assistingParticipantIds': int("".join(map(str, event['assistingParticipantIds']))) if 'assistingParticipantIds' in event else 0,
                'bounty': event['bounty'], 
                'killerId': event['killerId'],
                'killerTeamId': event['killerTeamId'],
                'monsterType': event['monsterType'],
                'timestamp': event['timestamp'],
            })
        elif event['type'] == 'GAME_END':
            game_end.append({
                'match_id': match_id,
                'gameId': event['gameId'],
                'realTimestamp': event['realTimestamp'],
                'timestamp': event['timestamp'],
                'winningTeam': event['winningTeam']
            })
        elif event['type'] == 'ITEM_DESTROYED':
            item_destroyed.append({
                'match_id': match_id,
                'itemId': event['itemId'],
                'participantId': event['participantId'],
                'timestamp': event['timestamp']
            })
        elif event['type'] == 'ITEM_PURCHASED':
            item_purchased.append({
                'match_id': match_id,
                'itemId': event['itemId'],
                'participantId': event['participantId'],
                'timestamp': event['timestamp']
            })
        elif event['type'] == 'ITEM_SOLD':
            item_sold.append({
                'match_id': match_id,
                'itemId': event['itemId'],
                'participantId' : event['participantId'],
                'timestamp' :event['timestamp']
            })
        elif event['type'] == 'LEVEL_UP':
            level_up.append({
                'match_id': match_id,
                "level": event['level'],
                "participantId": event['participantId'],
                "timestamp": event['timestamp']
            })
        elif event['type'] == 'OBJECTIVE_BOUNTY_PRESTART':
            objective_bounty_prestart.append({
                'match_id': match_id,
                "actualStartTime": event['actualStartTime'],
                "teamId": event['teamId'],
                "timestamp": event['timestamp'],
            })
        elif event['type'] == 'SKILL_LEVEL_UP':
            skill_level_up.append({
                'match_id': match_id,
                "levelUpType": event['levelUpType'],
                "participantId": event['participantId'],
                "skillSlot": event['skillSlot'],
                "timestamp": event['timestamp'],
            })
        elif event['type'] == 'TURRET_PLATE_DESTROYED':
            turret_plate_destroyed.append({
                'match_id': match_id,
                "killerId": event['killerId'],
                "laneType": event['laneType'],
                "teamId": event['teamId'],
                "timestamp": event['timestamp'],
            })
        elif event['type'] == 'WARD_KILL':
            ward_kill.append({
                'match_id': match_id,
                "killerId": event['killerId'],
                "timestamp": event['timestamp'],
                "wardType": event['wardType']
            })
        elif event['type'] == 'WARD_PLACED':
            ward_placed.append({
                'match_id': match_id,
                "creatorId": event['creatorId'],
                "timestamp": event['timestamp'],
                "wardType": event['wardType']
            })
champion_participant_id_df = pd.DataFrame(champion_participant_id)
champion_stat_per_timestamp_df =pd.DataFrame(champion_stat_per_timestamp)

building_kill_df = pd.DataFrame(building_kill)
champion_kill_df = pd.DataFrame(champion_kill)
champion_special_kill_df = pd.DataFrame(champion_special_kill)
elite_monster_kill_df = pd.DataFrame(elite_monster_kill)
game_end_df = pd.DataFrame(game_end)
item_destroyed_df = pd.DataFrame(item_destroyed)
item_purchased_df = pd.DataFrame(item_purchased)
item_sold_df = pd.DataFrame(item_sold)
level_up_df = pd.DataFrame(level_up)
objective_bounty_prestart_df= pd.DataFrame(objective_bounty_prestart)
skill_level_up_df = pd.DataFrame(skill_level_up)
turret_plate_destroyed_df = pd.DataFrame(turret_plate_destroyed)
ward_kill_df = pd.DataFrame(ward_kill)
ward_placed_df = pd.DataFrame(ward_placed)
```
